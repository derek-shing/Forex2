```
def get_data_and_save_to_s3(pairname,start,end):
    url="https://fxmarketapi.com/apitimeseries?api_key="+APIKey+\
    "&currency="+pairname+\
    "&start_date="+start+\
    "&end_date="+end+\
    "&format=ohlc"
    res= requests.get(url)
    print(res.content)
    if res.json():
      data = res.json()
      print(url,data)
      s3r.Object('forexderekshing', pairname+"\/"+start+".json").put(Body=json.dumps(data))

```
